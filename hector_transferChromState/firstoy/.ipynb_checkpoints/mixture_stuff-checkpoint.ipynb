# Discrete mixture stuff

Let's replicate the notebook for simple gaussian mixture model https://pyro.ai/examples/gmm.html


```python
import os
from collections import defaultdict
import torch
import numpy as np
import scipy.stats
from torch.distributions import constraints
from matplotlib import pyplot
%matplotlib inline

import pyro
import pyro.distributions as dist
from pyro import poutine
from pyro.infer.autoguide import AutoDelta
from pyro.optim import Adam
from pyro.infer import SVI, TraceEnum_ELBO, config_enumerate, infer_discrete
```

```python
K=2

@config_enumerate
def model(data):
    weights = pyro.sample('weights', dist.Dirichlet(0.5 * torch.ones(K)))
    scale = pyro.sample('scale', dist.LogNormal(0., 2.))
    with pyro.plate('components', K):
        locs = pyro.sample('locs', dist.Normal(0., 10.))
        
    with pyro.plate('data', len(data)):
        assignment = pyro.sample('assignment', dist.Categorical(weights))
        pyro.sample('obs', dist.Normal(locs[assignment], scale), obs=data)
```

```python
data = torch.tensor([0., 1., 10., 11., 12.])
```

```python
optim = pyro.optim.Adam({'lr': 0.1, 'betas': [0.8, 0.99]})
elbo = TraceEnum_ELBO(max_plate_nesting=1)
```

```python
def init_loc_fn(site):
    if site["name"] == "weights":
        return torch.ones(K) / K
    if site["name"] == "scale":
        return (data.var() / 2).sqrt()
    if site["name"] == "locs":
        return data[torch.multinomial(torch.ones(len(data)) / len(data), K)]
    raise ValueError(site["name"])
    
def initialize(seed):
    global global_guide, svi
    pyro.set_rng_seed(seed)
    pyro.clear_param_store()
    global_guide = AutoDelta(poutine.block(model, expose=['weights', 'locs', 'scale']),
                             init_loc_fn=init_loc_fn)
    svi = SVI(model, global_guide, optim, loss=elbo)
    return svi.loss(model, global_guide, data)

loss, seed = min((initialize(seed), seed) for seed in range(100))
initialize(seed)
print('seed={}, initial_loss={}'.format(seed, loss))
```

```python
from collections import defaultdict
gradient_norms = defaultdict(list)
for name, value in pyro.get_param_store().named_parameters():
    value.register_hook(lambda g, name=name: gradient_norms[name].append(g.norm().item()))
    
losses = []
for i in range(200):
    loss = svi.step(data)
    losses.append(loss)
    print('.' if i % 100 else '\n', end='')
```

```python
pyplot.figure(figsize=(10,3), dpi=100).set_facecolor('white')
pyplot.plot(losses)
pyplot.xlabel('iters')
pyplot.ylabel('loss')
pyplot.yscale('log')
pyplot.title('Convergence of SVI');
```

```python
pyplot.figure(figsize=(10,3), dpi=100).set_facecolor('white')
for name, gradient_norm in gradient_norms.items():
    pyplot.plot(gradient_norm, label=name)
pyplot.xlabel('iters')
pyplot.ylabel('gradient norm')
pyplot.yscale('log')
pyplot.legend(loc='best')
pyplot.title('Gradient norms');
```

```python
map_estimates=global_guide(data)
weights=map_estimates['weights']
locs=map_estimates['locs']
scale=map_estimates['scale']
print('weights={}'.format(weights.data.numpy()))
print('locs={}'.format(locs.data.numpy()))
print('scale={}'.format(scale.data.numpy()))
```

```python
X=np.arange(-3,15,0.1)
Y1=weights[0].item() * scipy.stats.norm.pdf((X - locs[0].item()) / scale.item())
Y2=weights[1].item() * scipy.stats.norm.pdf((X - locs[1].item()) / scale.item())

pyplot.figure(figsize=(10,4), dpi=100).set_facecolor('white')
pyplot.plot(X, Y1, 'r-')
pyplot.plot(X, Y2, 'b-')
pyplot.plot(X, Y1+Y2, 'k--')
pyplot.plot(data.data.numpy(), np.zeros(len(data)), 'k*')
pyplot.title('Density of two component GMM')
pyplot.ylabel('probability density');
```

```python
guide_trace = poutine.trace(global_guide).get_trace(data)
trained_model = poutine.replay(model, trace=guide_trace)

def classifier(data, temperature=0):
    inferred_model = infer_discrete(trained_model, temperature=temperature, first_available_dim=-2)
    trace = poutine.trace(inferred_model).get_trace(data)
    return trace.nodes["assignment"]["value"]
print(classifier(data))
```

```python
new_data = torch.arange(-3, 15, 0.1)
assignment = classifier(new_data)
pyplot.figure(figsize=(10,4), dpi=100).set_facecolor('white')
pyplot.plot(new_data.numpy(), assignment.numpy())
pyplot.title('MAP assignment')
pyplot.xlabel('data value')
pyplot.ylabel('class assignment');
```

```python
@config_enumerate
def full_guide(data):
    with poutine.block(hide_types=["param"]):
        global_guide(data)
        
    with pyro.plate('data', len(data)):
        assignment_probs = pyro.param("assignment_probs", torch.ones(len(data), K) / K,
                                      constraint=constraints.unit_interval)
        pyro.sample("assignment", dist.Categorical(assignment_probs))
```

```python
optim = pyro.optim.Adam({'lr': 0.2, 'betas': [0.8, 0.99]})
elbo = TraceEnum_ELBO(max_plate_nesting=1)
svi = SVI(model, full_guide, optim, loss=elbo)

gradient_norms = defaultdict(list)
svi.loss(model, full_guide, data)
for name, value in pyro.get_param_store().named_parameters():
    value.register_hook(lambda g, name=name: gradient_norms[name].append(g.norm().item()))
    
losses = []
for i in range(200):
    loss = svi.step(data)
    losses.append(loss)
    print('.' if i % 100 else '\n', end='')
```

```python
pyplot.figure(figsize=(10,3), dpi=100).set_facecolor('white')
pyplot.plot(losses)
pyplot.xlabel('iters')
pyplot.ylabel('loss')
pyplot.yscale('log')
pyplot.title('Convergence of SVI');
```

```python
pyplot.figure(figsize=(10,4), dpi=100).set_facecolor('white')
for name, grad_norms in gradient_norms.items():
    pyplot.plot(grad_norms, label=name)
pyplot.xlabel('iters')
pyplot.ylabel('gradient norm')
pyplot.yscale('log')
pyplot.legend(loc='best')
pyplot.title('Gradient norms during SVI');
```

```python
assignment_probs = pyro.param('assignment_probs')
pyplot.figure(figsize=(8, 3), dpi=100).set_facecolor('white')
pyplot.plot(data.data.numpy(), assignment_probs.data.numpy()[:, 0], 'ro',
            label='component with mean {:0.2g}'.format(locs[0]))
pyplot.plot(data.data.numpy(), assignment_probs.data.numpy()[:, 1], 'bo',
            label='component with mean {:0.2g}'.format(locs[1]))
pyplot.title('Mixture assignment probabilities')
pyplot.xlabel('data value')
pyplot.ylabel('assignment probability')
pyplot.legend(loc='center');
```

Now let's see about this paper: https://arxiv.org/pdf/1611.02648.pdf. Let's generate the syntetic dataset they are using

```python
import numpy as np
import math

def sample_arc(translation=2, rot_angle=0, num_points=1000):
    angles = (0.5 + torch.randn(num_points,1) * 0.15) * math.pi

    data = torch.cat((angles.cos(), angles.sin()),1) + torch.tensor([translation,0])
    z = torch.randn(*data.shape) * 0.05
    rot_matrix = torch.tensor([[math.cos(rot_angle), -math.sin(rot_angle)], 
                               [math.sin(rot_angle), math.cos(rot_angle)]])
    data = torch.matmul(rot_matrix, data.T).T + z
    return data
 
def generate_synthetic(translation=2, rot_angle=0, num_points=1000):
    data = torch.zeros((0,2))
    rot_angles = torch.arange(0,2,2/5) * math.pi

    for angle in rot_angles:
        data = torch.cat((data, sample_arc(rot_angle=angle)), axis=0)
    return data

data = generate_synthetic()
pyplot.plot(data[:,0], data[:,1], 'k.');
```

Let's repackage the GMM above, train it and see how it replicates fig 2.c


```python
import torch.nn as nn
import torch.nn.functional as F
from collections import defaultdict

class GMM(nn.Module):
    def __init__(self, input_dim=2, K=2):
        super().__init__()
        self.K = K
        self.input_dim = input_dim


    @config_enumerate
    def model(self, data):
        weights = pyro.sample('weights', dist.Dirichlet(0.5 * torch.ones(self.K)))
        with pyro.plate('components', self.K):
            locs = pyro.sample('locs', dist.Normal(torch.zeros((self.input_dim)), 10. * torch.ones((self.input_dim))).to_event(1))
            scales = pyro.sample('scales', dist.LogNormal(torch.zeros((self.input_dim)), 2. * torch.ones((self.input_dim))).to_event(1))
            
        with pyro.plate('data', len(data)):
            assignment = pyro.sample('assignment', dist.Categorical(weights))
            pyro.sample('obs', dist.Normal(locs[assignment], scales[assignment]).to_event(1), obs=data)
            
    def guide(self, data):
        return AutoDelta(poutine.block(self.model, 
            expose=['weights', 'locs', 'scales']),
            init_loc_fn=self.init_loc_fn(data))
    
    def compute_scales(self, data, locs):
        N = data.shape[0]
        
        weights = torch.zeros((N, self.K))
        for k in range(self.K):
            weights[:,k] = torch.sum((data - locs[k,:])**2, axis=1).sqrt()
        
        scales = torch.zeros((self.K, self.input_dim))
        for k in range(K):
            scales[k,:] = torch.sqrt(torch.sum(weights[:,k].repeat(2,1).T * (data - locs[k,:]) ** 2, 0) / torch.sum(weights[:,k]))
        
        return scales
        
    def init_loc_fn(self, data):
        nrows = data.shape[0]
        draw = torch.multinomial(torch.ones(nrows) / nrows, self.K)
        locs = data[draw,:]
        var_samples = torch.multinomial(torch.ones(nrows) / nrows, 100)
        scales = self.compute_scales(data[var_samples,:], locs)
        
        def _f(site):
            if site["name"] == "weights":
                return torch.ones(self.K) / self.K
            if site["name"] == "scales":
                return scales
            if site["name"] == "locs":
                return locs
            raise ValueError(site["name"])
        return _f
    
class Trainer:
    def __init__(self, module, data, batch_size=30):
        self.module = module
        self.data = data
        self.optim = pyro.optim.Adam({'lr': 0.1, 'betas': [0.8, 0.99]})
        self.elbo = TraceEnum_ELBO(max_plate_nesting=1)
        self.guide = module.guide(data)
        self.svi = SVI(module.model, self.guide, self.optim, loss=self.elbo)
        self.loader = torch.utils.data.DataLoader(self.data, batch_size=batch_size, shuffle=True)
        
    def _initialize(self, seed):
        pyro.set_rng_seed(seed)
        pyro.clear_param_store()
        return self.svi.loss(self.module.model, self.guide, self.data)
    
    def initialize(self):
        loss, seed = min((self._initialize(seed), seed) for seed in range(100))
        self._initialize(seed)
        print('seed={}, initial_loss={}'.format(seed, loss))
   
    def train_step(self):
        epoch_loss = 0.
        for x in self.loader:
            epoch_loss += self.svi.step(x)
        
        return epoch_loss / len(self.data)

    def train(self, initialize=True, num_epochs=200, print_step=5):
        if initialize:
            self.initialize()
        gradient_norms = defaultdict(list)
        for name, value in pyro.get_param_store().named_parameters():
            value.register_hook(lambda g, name=name: gradient_norms[name].append(g.norm().item()))
    
        losses = []
        for i in range(num_epochs):
            loss = self.train_step()
            losses.append(loss)
            if i % print_step == 0:
                print('epoch {:d}: {:.3e}'.format(i, loss))
                
        self.losses = losses
        self.gradient_norms = gradient_norms
        return losses, gradient_norms
    
    def loss_plot(self):
        pyplot.figure(figsize=(10,3), dpi=100).set_facecolor('white')
        pyplot.plot(losses)
        pyplot.xlabel('iters')
        pyplot.ylabel('loss')
        pyplot.yscale('log')
        pyplot.title('Convergence of SVI')
        pyplot.show();
        
    def grad_plot(self):
        pyplot.figure(figsize=(10,3), dpi=100).set_facecolor('white')
        for name, gradient_norm in gradient_norms.items():
            pyplot.plot(gradient_norm, label=name)
        pyplot.xlabel('iters')
        pyplot.ylabel('gradient norm')
        pyplot.yscale('log')
        pyplot.legend(loc='best')
        pyplot.title('Gradient norms')
        pyplot.show();
        
    def make_plots(self):
        self.loss_plot()
        self.grad_plot()
        
    def get_estimates(self):
        torch_estimates = self.guide(self.data)
        estimates = {}
        for name, tensor in torch_estimates.items():
            estimates[name] = tensor.data.numpy()
        return estimates
```

```python
gmm = GMM(K=5)
trainer = Trainer(gmm, data)
trainer.train(num_epochs=20);
trainer.make_plots()
trainer.get_estimates()
```
